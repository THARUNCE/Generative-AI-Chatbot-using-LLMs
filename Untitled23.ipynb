```python
# CELL 2 - Imports and Configuration
import os, time, warnings
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
warnings.filterwarnings('ignore')

from torch.nn import Linear, BatchNorm1d
from torch_geometric.datasets import MalNetTiny
from torch_geometric.loader import DataLoader
from torch_geometric.nn import (GCNConv, GATConv, SAGEConv, GINConv,
                                 global_mean_pool, global_add_pool)
from torch_geometric.transforms import LocalDegreeProfile
from sklearn.metrics import (accuracy_score, precision_score,
                              recall_score, f1_score,
                              confusion_matrix, classification_report)

os.makedirs('results', exist_ok=True)

DEVICE      = torch.device('cpu')
HIDDEN_DIM  = 64
NUM_CLASSES = 5
EPOCHS      = 15
BATCH_SIZE  = 64
CLASS_NAMES = ['AdDisplay', 'Adware', 'Benign', 'Downloader', 'Trojan']
COLORS      = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12', '#9b59b6']
IN_FEATURES = 5

print('Setup complete!')
print('Device  : ' + str(DEVICE))
print('Epochs  : ' + str(EPOCHS))
print('PyTorch : ' + torch.__version__)
```

```python
# CELL 3 - Load Dataset
# First time downloads MalNet-Tiny (~2-3 mins)
print('Loading MalNet-Tiny dataset...')
transform = LocalDegreeProfile()
dataset = MalNetTiny(root='data/MalNetTiny')
dataset = dataset.shuffle()

total     = len(dataset)
train_end = int(0.70 * total)
val_end   = int(0.85 * total)

print('Applying node features...')
train_dataset = [transform(d) for d in dataset[:train_end]]
val_dataset   = [transform(d) for d in dataset[train_end:val_end]]
test_dataset  = [transform(d) for d in dataset[val_end:]]

train_loader = DataLoader(train_dataset[:1200], batch_size=64, shuffle=True)
val_loader   = DataLoader(val_dataset[:400],    batch_size=64, shuffle=False)
test_loader  = DataLoader(test_dataset[:400],   batch_size=64, shuffle=False)

print('Dataset loaded!')
print('Total graphs : ' + str(total))
print('Train        : 1200')
print('Validation   : 400')
print('Test         : 400')
print('Node features: 5 (degree-based)')
```

```python
# CELL 4 - Define All 4 GNN Models
class GCN(torch.nn.Module):
    def __init__(self, in_ch, hidden, out_ch):
        super().__init__()
        self.conv1 = GCNConv(in_ch, hidden)
        self.bn1   = BatchNorm1d(hidden)
        self.conv2 = GCNConv(hidden, hidden)
        self.bn2   = BatchNorm1d(hidden)
        self.conv3 = GCNConv(hidden, hidden)
        self.bn3   = BatchNorm1d(hidden)
        self.lin   = Linear(hidden, out_ch)
    def forward(self, x, edge_index, batch):
        x = F.relu(self.bn1(self.conv1(x, edge_index)))
        x = F.dropout(x, p=0.3, training=self.training)
        x = F.relu(self.bn2(self.conv2(x, edge_index)))
        x = F.dropout(x, p=0.3, training=self.training)
        x = F.relu(self.bn3(self.conv3(x, edge_index)))
        x = global_mean_pool(x, batch)
        return self.lin(x)

class GAT(torch.nn.Module):
    def __init__(self, in_ch, hidden, out_ch, heads=4):
        super().__init__()
        self.conv1 = GATConv(in_ch, hidden, heads=heads, dropout=0.3)
        self.conv2 = GATConv(hidden*heads, hidden, heads=heads, dropout=0.3)
        self.conv3 = GATConv(hidden*heads, hidden, heads=1, concat=False, dropout=0.3)
        self.lin   = Linear(hidden, out_ch)
    def forward(self, x, edge_index, batch):
        x = F.elu(self.conv1(x, edge_index))
        x = F.dropout(x, p=0.3, training=self.training)
        x = F.elu(self.conv2(x, edge_index))
        x = F.dropout(x, p=0.3, training=self.training)
        x = F.elu(self.conv3(x, edge_index))
        x = global_mean_pool(x, batch)
        return self.lin(x)

class GraphSAGE(torch.nn.Module):
    def __init__(self, in_ch, hidden, out_ch):
        super().__init__()
        self.conv1 = SAGEConv(in_ch, hidden)
        self.bn1   = BatchNorm1d(hidden)
        self.conv2 = SAGEConv(hidden, hidden)
        self.bn2   = BatchNorm1d(hidden)
        self.conv3 = SAGEConv(hidden, hidden)
        self.bn3   = BatchNorm1d(hidden)
        self.lin   = Linear(hidden, out_ch)
    def forward(self, x, edge_index, batch):
        x = F.relu(self.bn1(self.conv1(x, edge_index)))
        x = F.dropout(x, p=0.3, training=self.training)
        x = F.relu(self.bn2(self.conv2(x, edge_index)))
        x = F.dropout(x, p=0.3, training=self.training)
        x = F.relu(self.bn3(self.conv3(x, edge_index)))
        x = global_mean_pool(x, batch)
        return self.lin(x)

class GIN(torch.nn.Module):
    def __init__(self, in_ch, hidden, out_ch):
        super().__init__()
        def mlp(i, o):
            return torch.nn.Sequential(
                Linear(i, o), BatchNorm1d(o),
                torch.nn.ReLU(), Linear(o, o))
        self.conv1 = GINConv(mlp(in_ch, hidden))
        self.conv2 = GINConv(mlp(hidden, hidden))
        self.conv3 = GINConv(mlp(hidden, hidden))
        self.lin1  = Linear(hidden, hidden)
        self.lin2  = Linear(hidden, out_ch)
    def forward(self, x, edge_index, batch):
        x = F.relu(self.conv1(x, edge_index))
        x = F.dropout(x, p=0.3, training=self.training)
        x = F.relu(self.conv2(x, edge_index))
        x = F.dropout(x, p=0.3, training=self.training)
        x = F.relu(self.conv3(x, edge_index))
        x = global_add_pool(x, batch)
        x = F.relu(self.lin1(x))
        x = F.dropout(x, p=0.5, training=self.training)
        return self.lin2(x)

print('All 4 models defined!')
print('GCN       : 3 layers + BatchNorm + Dropout')
print('GAT       : 3 layers + 4 attention heads')
print('GraphSAGE : 3 layers + BatchNorm + Dropout')
print('GIN       : 3 layers + MLP aggregation')
```

```python
# CELL 5 - Training and Evaluation Functions
def train_epoch(model, loader, optimizer, criterion):
    model.train()
    total_loss = 0
    for data in loader:
        data = data.to(DEVICE)
        optimizer.zero_grad()
        out  = model(data.x, data.edge_index, data.batch)
        loss = criterion(out, data.y)
        loss.backward()
        optimizer.step()
        total_loss += loss.item() * data.num_graphs
    return total_loss / len(loader.dataset)

def evaluate(model, loader):
    model.eval()
    preds, labels = [], []
    with torch.no_grad():
        for data in loader:
            data = data.to(DEVICE)
            out  = model(data.x, data.edge_index, data.batch)
            pred = out.argmax(dim=1)
            preds.extend(pred.cpu().numpy())
            labels.extend(data.y.cpu().numpy())
    return accuracy_score(labels, preds), preds, labels

def train_model(model, name):
    print('Training ' + name + '...')
    model     = model.to(DEVICE)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
    scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.5)
    criterion = torch.nn.CrossEntropyLoss()
    best_val, best_state = 0, None
    start = time.time()

    for epoch in range(1, EPOCHS + 1):
        loss    = train_epoch(model, train_loader, optimizer, criterion)
        val_acc, _, _ = evaluate(model, val_loader)
        scheduler.step()
        if val_acc > best_val:
            best_val   = val_acc
            best_state = {k: v.clone() for k, v in model.state_dict().items()}
        print('Epoch ' + str(epoch) + '/' + str(EPOCHS) +
              '  Loss: ' + str(round(loss, 4)) +
              '  Val: '  + str(round(val_acc, 4)))

    model.load_state_dict(best_state)
    train_time = time.time() - start

    test_acc, preds, labels = evaluate(model, test_loader)
    prec = precision_score(labels, preds, average='weighted', zero_division=0)
    rec  = recall_score(labels, preds,    average='weighted', zero_division=0)
    f1   = f1_score(labels, preds,        average='weighted', zero_division=0)

    print('-------------------------------')
    print('RESULTS : ' + name)
    print('Accuracy  : ' + str(round(test_acc * 100, 2)) + '%')
    print('Precision : ' + str(round(prec, 4)))
    print('Recall    : ' + str(round(rec, 4)))
    print('F1 Score  : ' + str(round(f1, 4)))
    print('Time      : ' + str(round(train_time, 1)) + ' seconds')
    print('-------------------------------')

    cm = confusion_matrix(labels, preds)
    cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_norm, annot=True, fmt='.2f', cmap='Blues',
                xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES)
    plt.title(name + ' - Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.xticks(rotation=30, ha='right')
    plt.tight_layout()
    plt.savefig('results/' + name + '_confusion_matrix.png', dpi=150, bbox_inches='tight')
    plt.show()
    print(classification_report(labels, preds, target_names=CLASS_NAMES, zero_division=0))

    return {'model': model, 'name': name, 'accuracy': test_acc,
            'precision': prec, 'recall': rec, 'f1': f1,
            'time': train_time, 'preds': preds, 'labels': labels}

print('Functions ready!')
print('Now run Cell 6 to start training GCN')
```

```python
# CELL 6 - Train GCN (~5-8 mins on CPU)
gcn_result = train_model(GCN(IN_FEATURES, HIDDEN_DIM, NUM_CLASSES), 'GCN')
```

```python
# CELL 7 - Train GAT (~8-10 mins on CPU)
gat_result = train_model(GAT(IN_FEATURES, HIDDEN_DIM, NUM_CLASSES), 'GAT')
```

```python
# CELL 8 - Train GraphSAGE (~5-8 mins on CPU)
sage_result = train_model(GraphSAGE(IN_FEATURES, HIDDEN_DIM, NUM_CLASSES), 'GraphSAGE')
```

```python
# CELL 9 - Train GIN (~5-8 mins on CPU)
gin_result = train_model(GIN(IN_FEATURES, HIDDEN_DIM, NUM_CLASSES), 'GIN')
```

```python
# CELL 11 - GNNExplainer on best model (GraphSAGE)
from torch_geometric.explain import Explainer, GNNExplainer

print('Running GNNExplainer on GraphSAGE (best model)...')

# Get one sample per class from test set
samples = {}
for data in test_dataset[:400]:
    label = data.y.item()
    if label not in samples and data.num_nodes >= 5:
        samples[label] = data
    if len(samples) == NUM_CLASSES:
        break

# Build explainer
best_model = sage_result['model']
best_model.eval()

explainer = Explainer(
    model=best_model,
    algorithm=GNNExplainer(epochs=100),
    explanation_type='model',
    node_mask_type='attributes',
    edge_mask_type='object',
    model_config=dict(
        mode='multiclass_classification',
        task_level='graph',
        return_type='raw'
    )
)

# Plot explanation for each class
fig, axes = plt.subplots(1, len(samples), figsize=(5*len(samples), 5))
if len(samples) == 1:
    axes = [axes]

for idx, (label, data) in enumerate(samples.items()):
    data = data.to(DEVICE)
    exp  = explainer(
        x=data.x,
        edge_index=data.edge_index,
        batch=torch.zeros(data.num_nodes, dtype=torch.long, device=DEVICE)
    )
    mask = exp.edge_mask.cpu().detach().numpy()
    top  = np.argsort(mask)[-15:]
    axes[idx].bar(range(len(top)), mask[top], color='#e74c3c', alpha=0.8)
    axes[idx].set_title(CLASS_NAMES[label], fontsize=11, fontweight='bold')
    axes[idx].set_xlabel('Top 15 edges')
    axes[idx].set_ylabel('Importance score')
    axes[idx].grid(axis='y', alpha=0.3)
    print(CLASS_NAMES[label] + ' - max importance: ' + str(round(mask.max(), 4)))

fig.suptitle('GNNExplainer - Important Subgraph Edges per Malware Family (GraphSAGE)',
             fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('results/gnnexplainer_results.png', dpi=150, bbox_inches='tight')
plt.show()
print('GNNExplainer complete! Saved to results/gnnexplainer_results.png')
```

```python
# CELL 13 - ROC Curves + Per-Class F1 Deep Evaluation
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc

print('Running deep evaluation...')

# Store all results manually using your training results
all_results = [
    {'name': 'GCN',       'accuracy': 0.66, 'f1': 0.63},
    {'name': 'GAT',       'accuracy': 0.70, 'f1': 0.68},
    {'name': 'GraphSAGE', 'accuracy': 0.71, 'f1': 0.70},
    {'name': 'GIN',       'accuracy': 0.54, 'f1': 0.51},
]

# Plot 1 - Per class F1 heatmap using GraphSAGE (best model)
print('Getting per-class predictions from GraphSAGE...')
best_model = sage_result['model']
best_model.eval()

all_preds, all_labels, all_probs = [], [], []
with torch.no_grad():
    for data in test_loader:
        data  = data.to(DEVICE)
        out   = best_model(data.x, data.edge_index, data.batch)
        probs = torch.softmax(out, dim=1)
        pred  = out.argmax(dim=1)
        all_preds.extend(pred.cpu().numpy())
        all_labels.extend(data.y.cpu().numpy())
        all_probs.extend(probs.cpu().numpy())

all_probs  = np.array(all_probs)
all_labels = np.array(all_labels)
all_preds  = np.array(all_preds)

# Per class F1 bar chart
from sklearn.metrics import f1_score
per_class_f1 = f1_score(all_labels, all_preds, average=None, zero_division=0)
fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.bar(CLASS_NAMES, per_class_f1, color=COLORS, alpha=0.85, edgecolor='white')
for bar, val in zip(bars, per_class_f1):
    ax.text(bar.get_x() + bar.get_width()/2,
            bar.get_height() + 0.01,
            str(round(val, 3)), ha='center',
            fontweight='bold', fontsize=11)
ax.set_title('Per-Class F1 Score - GraphSAGE (Best Model)',
             fontsize=14, fontweight='bold')
ax.set_ylabel('F1 Score')
ax.set_ylim(0, 1.1)
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('results/04_per_class_f1.png', dpi=150, bbox_inches='tight')
plt.show()

# ROC Curves - one per class
labels_bin = label_binarize(all_labels, classes=list(range(NUM_CLASSES)))
fig, ax = plt.subplots(figsize=(10, 7))
for i in range(NUM_CLASSES):
    fpr, tpr, _ = roc_curve(labels_bin[:, i], all_probs[:, i])
    roc_auc     = auc(fpr, tpr)
    ax.plot(fpr, tpr, color=COLORS[i], linewidth=2,
            label=CLASS_NAMES[i] + ' (AUC = ' + str(round(roc_auc, 3)) + ')')

ax.plot([0, 1], [0, 1], 'k--', linewidth=1)
ax.set_title('ROC Curves per Malware Family - GraphSAGE',
             fontsize=14, fontweight='bold')
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
ax.legend(loc='lower right', fontsize=11)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('results/05_roc_curves.png', dpi=150, bbox_inches='tight')
plt.show()

# Model accuracy comparison bar
fig, ax = plt.subplots(figsize=(9, 5))
names = [r['name'] for r in all_results]
accs  = [r['accuracy'] for r in all_results]
f1s   = [r['f1'] for r in all_results]
x = np.arange(len(names))
w = 0.35
b1 = ax.bar(x - w/2, accs, w, label='Accuracy', color='#3498db', alpha=0.85)
b2 = ax.bar(x + w/2, f1s,  w, label='F1 Score', color='#e74c3c', alpha=0.85)
for bar in list(b1) + list(b2):
    ax.text(bar.get_x() + bar.get_width()/2,
            bar.get_height() + 0.01,
            str(round(bar.get_height(), 2)),
            ha='center', fontweight='bold', fontsize=10)
ax.set_xticks(x)
ax.set_xticklabels(names, fontsize=12)
ax.set_ylim(0, 1.0)
ax.set_title('Accuracy vs F1 Score - All Models',
             fontsize=14, fontweight='bold')
ax.set_ylabel('Score')
ax.legend(fontsize=11)
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('results/06_accuracy_vs_f1.png', dpi=150, bbox_inches='tight')
plt.show()
print('Deep evaluation complete! 3 plots saved.')
```

```python
# CELL 14 - Ablation Study (GraphSAGE - best model)
# Tests effect of: number of layers, hidden dimensions, epochs
print('Running ablation study...')
print('This will take 15-20 mins on CPU - please wait...')

def quick_train(model, epochs=10):
    model = model.to(DEVICE)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
    criterion = torch.nn.CrossEntropyLoss()
    for epoch in range(epochs):
        model.train()
        for data in train_loader:
            data = data.to(DEVICE)
            optimizer.zero_grad()
            out  = model(data.x, data.edge_index, data.batch)
            loss = criterion(out, data.y)
            loss.backward()
            optimizer.step()
    model.eval()
    preds, labels = [], []
    with torch.no_grad():
        for data in test_loader:
            data = data.to(DEVICE)
            out  = model(data.x, data.edge_index, data.batch)
            pred = out.argmax(dim=1)
            preds.extend(pred.cpu().numpy())
            labels.extend(data.y.cpu().numpy())
    return accuracy_score(labels, preds)

# Ablation 1 - Number of layers
print('Testing number of layers...')

class SAGE1Layer(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = SAGEConv(IN_FEATURES, 64)
        self.lin   = Linear(64, NUM_CLASSES)
    def forward(self, x, edge_index, batch):
        x = F.relu(self.conv1(x, edge_index))
        x = global_mean_pool(x, batch)
        return self.lin(x)

class SAGE2Layer(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = SAGEConv(IN_FEATURES, 64)
        self.conv2 = SAGEConv(64, 64)
        self.lin   = Linear(64, NUM_CLASSES)
    def forward(self, x, edge_index, batch):
        x = F.relu(self.conv1(x, edge_index))
        x = F.relu(self.conv2(x, edge_index))
        x = global_mean_pool(x, batch)
        return self.lin(x)

class SAGE3Layer(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = SAGEConv(IN_FEATURES, 64)
        self.conv2 = SAGEConv(64, 64)
        self.conv3 = SAGEConv(64, 64)
        self.lin   = Linear(64, NUM_CLASSES)
    def forward(self, x, edge_index, batch):
        x = F.relu(self.conv1(x, edge_index))
        x = F.relu(self.conv2(x, edge_index))
        x = F.relu(self.conv3(x, edge_index))
        x = global_mean_pool(x, batch)
        return self.lin(x)

layer_accs = []
for name, model in [('1 Layer', SAGE1Layer()),
                     ('2 Layers', SAGE2Layer()),
                     ('3 Layers', SAGE3Layer())]:
    acc = quick_train(model, epochs=10)
    layer_accs.append(acc)
    print(name + ' : ' + str(round(acc * 100, 2)) + '%')

# Ablation 2 - Hidden dimensions
print('Testing hidden dimensions...')

class SAGEHidden(torch.nn.Module):
    def __init__(self, hidden):
        super().__init__()
        self.conv1 = SAGEConv(IN_FEATURES, hidden)
        self.conv2 = SAGEConv(hidden, hidden)
        self.conv3 = SAGEConv(hidden, hidden)
        self.lin   = Linear(hidden, NUM_CLASSES)
    def forward(self, x, edge_index, batch):
        x = F.relu(self.conv1(x, edge_index))
        x = F.relu(self.conv2(x, edge_index))
        x = F.relu(self.conv3(x, edge_index))
        x = global_mean_pool(x, batch)
        return self.lin(x)

hidden_dims = [32, 64, 128]
hidden_accs = []
for h in hidden_dims:
    acc = quick_train(SAGEHidden(h), epochs=10)
    hidden_accs.append(acc)
    print('Hidden ' + str(h) + ' : ' + str(round(acc * 100, 2)) + '%')

# Ablation 3 - Number of epochs
print('Testing number of epochs...')
epoch_list = [5, 10, 15]
epoch_accs = []
for e in epoch_list:
    acc = quick_train(SAGEHidden(64), epochs=e)
    epoch_accs.append(acc)
    print('Epochs ' + str(e) + ' : ' + str(round(acc * 100, 2)) + '%')

# Plot all 3 ablation studies
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Ablation Study - GraphSAGE', fontsize=15, fontweight='bold')

axes[0].bar(['1 Layer', '2 Layers', '3 Layers'],
            [a*100 for a in layer_accs],
            color=['#3498db', '#2ecc71', '#e74c3c'], alpha=0.85)
for i, v in enumerate(layer_accs):
    axes[0].text(i, v*100 + 0.3, str(round(v*100, 1)) + '%',
                 ha='center', fontweight='bold')
axes[0].set_title('Effect of Number of Layers')
axes[0].set_ylabel('Accuracy (%)')
axes[0].set_ylim(0, 100)
axes[0].grid(axis='y', alpha=0.3)

axes[1].bar([str(h) for h in hidden_dims],
            [a*100 for a in hidden_accs],
            color=['#3498db', '#2ecc71', '#e74c3c'], alpha=0.85)
for i, v in enumerate(hidden_accs):
    axes[1].text(i, v*100 + 0.3, str(round(v*100, 1)) + '%',
                 ha='center', fontweight='bold')
axes[1].set_title('Effect of Hidden Dimensions')
axes[1].set_ylabel('Accuracy (%)')
axes[1].set_ylim(0, 100)
axes[1].grid(axis='y', alpha=0.3)

axes[2].plot([str(e) for e in epoch_list],
             [a*100 for a in epoch_accs],
             marker='o', color='#9b59b6', linewidth=2, markersize=8)
for i, v in enumerate(epoch_accs):
    axes[2].text(i, v*100 + 0.3, str(round(v*100, 1)) + '%',
                 ha='center', fontweight='bold')
axes[2].set_title('Effect of Training Epochs')
axes[2].set_ylabel('Accuracy (%)')
axes[2].set_ylim(0, 100)
axes[2].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('results/07_ablation_study.png', dpi=150, bbox_inches='tight')
plt.show()
print('Ablation study complete! Saved to results/07_ablation_study.png')
```
